```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('cereal.csv', index_col=0)
# print(df.head())
ax = df.plot(x='carbo', y='calories', kind='scatter', title='Carbohydrates vs. Calories')
ax.set_xlabel("Carbohydrates per Cup (g)")
ax.set_ylabel("Calories per Cup (KCal)")
plt.show()
```

```python
ax = df.plot(x='fat', y='calories', kind='scatter', title='Fat vs. Calories')
ax.set_xlabel("Fat per Cup (g)")
ax.set_ylabel("Calories per Cup (KCal)")
plt.show()
```
